```python
# Importing the required libraries
from pyvis import network as net
#net = Network()

import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
G=nx.Graph()

from datetime import date
today = date.today()

print(today)
```

```python
# Loading data into pandas dataframe form .CSV file -----

df_billions = pd.read_csv('https://raw.githubusercontent.com/tagensingh/DATA620-WEEK-3/main/billionaires_sample2.csv')

df_billions.head(20)
```

```python
G.clear()
G = nx.from_pandas_edgelist(df_billions, source='name', target='citizenship', create_using=nx.Graph())
G = nx.from_pandas_edgelist(df_billions, source='name', target='industries', create_using=nx.Graph())
G = nx.from_pandas_edgelist(df_billions, source='citizenship', target='industries', create_using=nx.Graph())
```

```python
print (nx.info(G))
```

```python
nx.draw_networkx_nodes(G,pos=nx.spring_layout(G),node_color='r')

# edges
nx.draw_networkx_edges(G,pos=nx.spring_layout(G),alpha=0.5,width=20,edge_color='k', style='solid',)

# labels
nx.draw_networkx_labels(G,pos=nx.spring_layout(G),font_size=8,font_family='sans-serif')

plt.axis()
plt.show() # display
```
